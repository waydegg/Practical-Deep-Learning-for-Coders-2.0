# Pets

```python
from fastai2.basics import *
from fastai2.vision.all import *
from fastai2.callback.all import *
```

```python
torch.cuda.set_device(2)
```

```python
# set seed
np.random.seed(2)
```

Building a DataBunch

```python
path = untar_data(URLs.PETS)
fnames = get_image_files(path/'images')
pat = r'(.+)_\d+.jpg'
```

```python
path.ls()
```

Set basic transforms

```python
batch_tfms = [*aug_transforms(size=224, max_warp=0), Normalize.from_stats(*imagenet_stats)]
item_tfms = RandomResizedCrop(460, min_scale=0.75, ratio=(1.,1.))
bs = 64
```

Create a `ImageDataLoader` directly from the class

```python
dls = ImageDataLoaders.from_name_re(path, fnames, pat, batch_tfms=batch_tfms, item_tfms=item_tfms)
```

Now recreate this DataLoader object starting from a `DataBlock`. Then we will create a `DataLoader` from this datablock.

```python
pets = DataBlock(blocks=(ImageBlock, CategoryBlock),
                get_items=get_image_files,
                splitter=RandomSplitter(),
                get_y=RegexLabeller(pat=r'/([^/]+)_\d+.*'),
                item_tfms=item_tfms,
                batch_tfms=batch_tfms)
```

```python
path_im = path/'images'
```

```python
dls = pets.dataloaders(path_im, bs=bs)
```

```python
dls.show_batch(max_n=9, figsize=(6,7))
```

Lets look at the `vocab`, or *labels/calsses* in our data. Returned to us is a `CategoryMap`, a new way of storing data (hence the '#37' at the beginning of the vocab, denoting the number of classes)

```python
dls.vocab
```

`o2i` is object-to-item

```python
dls.vocab.o2i
```

### Model Training

`.to_fp16()` is using `Mixed Precision` during model training to speed it up. The __loss function__ and __optimizer__ are infered from the `DataLoader`, so we don't need to define them here.

```python
learn = cnn_learner(dls, resnet34, pretrained=True, metrics=error_rate).to_fp16()
```

```python
learn.opt_func, learn.loss_func
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(4, lr_max=slice(1e-6, 1e-5))
```

### Look at losses

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
losses,idxs = interp.top_losses()

len(dls.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,10))
```

```python
interp.plot_confusion_matrix(figsize=(12,12))
```
